```python
input = '''L68
L30
R48
L5
R60
L55
L1
L99
R14
L82'''
#input = open('.day01','r').read().strip()
```

```python
from operator import add, sub
rots = [ (add if x[0] == 'R' else sub,int(x[1:])) for x in input.split('\n') ]
```

```python
i = 50
at_zero = 0
for op,n in rots:
    i = op(i, n) % 100
    if i == 0:
        at_zero += 1
at_zero
```

```python
i = 50
at_zero = 0
for op,n in rots:
    # idk just do it by hand
    for _ in range(n):
        i = op(i, 1) % 100
        if i == 0:
            at_zero += 1
at_zero
```

```python
from itertools import accumulate
d = input

sum([y==0 for y in accumulate([int(x[1:])*(-1if x[0]=='L'else 1) for x in d.split('\n')], lambda k,j: (k+j)%100,initial=50)])
```
