```python
input = open('.day01','r').read().strip()
```

```python
from operator import add, sub
rots = [ (add if x[0] == 'R' else sub,int(x[1:])) for x in input.split('\n') ]
```

```python
i = 50
at_zero = 0
for op,n in rots:
    i = op(i, n) % 100
    if i == 0:
        at_zero += 1
at_zero
```
