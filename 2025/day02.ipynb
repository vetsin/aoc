```python
def is_repeating(num: int):
    s = str(num)
    if len(s) <= 1 or len(s)%2 != 0:
        return False
    mid = len(s) // 2
    return s[:mid] == s[mid:]
```

```python
from textwrap import wrap
def is_repeating(num: int):
    s = str(num)
    return any(len(set(wrap(s,i))) == 1 for i in range(len(s)//2,0,-1))

sum([i for (low,high) in products for i in range(low,high+1) if is_repeating(i)])
```
