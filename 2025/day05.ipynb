```python
from sympy import Interval, Union
def union(data):
    """ Union of a list of intervals e.g. [(1,2),(3,4)] """
    intervals = [Interval(begin, end) for (begin, end) in data]
    u = Union(*intervals)
    return [list(u.args[:2])] if isinstance(u, Interval) \
       else list(u.args)
```

```python
ggoods = [ (int(i),int(j)) for (i,j) in [ x.split('-') for x in goods_str.split('\n') ]]
ggoods = union(ggoods)
```

```python
t = 0
for g in ggoods:
    if type(g) == Interval:
        s = (g.end+1) - g.start
        t += s
    else:
        t += len(g)
print(t)
```
