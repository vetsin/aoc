```python
inp = '''987654321111111
811111111111119
234234234234278
818181911112111'''
inp = open('untitled.txt','r').read()
batts = inp.split('\n')
```

```python
s = 0
for b in batts:
    l = b.index(max(b[:-1]))
    r = b.index(max(b[l+1:]))
    i = b[l]+b[r]
    s += int(i)
s
```

```python
def max_index(values):
    return max(range(len(values)), key=values.__getitem__)

s = 0
for b in batts:
    idxs = [ max_index(b[:-11]) ]
    while len(idxs) < 12:
        last = idxs[-1]+1
        until = len(b)-11+len(idxs)
        idxs.append(max_index(b[last:until])+last)
    s += int(''.join([b[i] for i in idxs]))
s
```
