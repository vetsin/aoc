```python
import operator
from functools import reduce
import numpy as np
from itertools import tee
```

```python
data = '''123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  '''
```

```python
from aocd import get_data
data = get_data(day=6, year=2025)
```

```python
lines = np.array([x.split() for x in data.split('\n')])
```

```python
ops = {'*':operator.mul,'+':operator.add}
sum(reduce(ops[l[-1]], [int(i) for i in l[:-1]]) for l in lines.T)
```

```python
# gross, whatever
lines = data.split('\n')
splits = [0]+[i for i in range(len(lines[0])) if all(line[i] == ' ' for line in lines)]+[len(lines[0])]
mops = iter([ops[l] for l in lines[-1].split()])
r = []
for s in lines[:-1]:
    start, end = tee(splits)
    next(end)
    r.append([s[i:j][::-1] for i,j in zip(start, end)])
total = 0
for row in np.array(r).T:
    rnums = []
    for nums in np.array([list(e) for e in row]).T:
        s = ''.join(nums).strip()
        if s:
            rnums.append(int(s))
    total += reduce(next(mops),rnums)
total
```
