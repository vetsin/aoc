```python
with open('day10.txt', 'r') as f:
    input = f.read()
```

```python
r = dict(noop=1, addx=2)
registers = dict(x=1)
cycle = 0

log = {}
for line in [x.split() for x in input.split('\n')]:
    command = line[0]
    counts = list(range(r[command]))
    for c in counts:
        cycle += 1
        log[cycle] = dict(registers)
        
    if command.startswith("add"):
        register = command[3]
        registers[register] += int(line[1])
```

```python
sum([ log[x]['x']*x for x in [20, 60, 100, 140, 180, 220] ])
```

```python
for i in range(6):
    line = [' ']*40
    for j in range(40):
        sprite = log[(i*40)+j+1]['x']
        if j in [sprite-1, sprite, sprite+1]:
            line[j] = '#'
    print(''.join(line))
```
