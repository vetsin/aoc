```python
with open('day9.txt', 'r') as f:
    input = f.read()
```

```python
moves = [ x.split() for x in input.split('\n') ]
moves = [ [x, int(y)] for x, y in moves ]
```

```python
def move(pos, direction, distance=1):
    if direction == "R":
        pos[0] += distance
    elif direction == "L":
        pos[0] -= distance
    elif direction == "U":
        pos[1] += distance
    elif direction == "D":
        pos[1] -= distance
    return pos
```

```python
def tail_move(head, tail):
    dx = head[0] - tail[0]
    dy = head[1] - tail[1]
    if dx > 1:
        tail[0] += 1
        tail[1] = head[1]
    if dx < -1:
        tail[0] -= 1
        tail[1] = head[1]
    if dy > 1:
        tail[1] += 1
        tail[0] = head[0]
    if dy < -1:
        tail[1] -= 1
        tail[0] = head[0]
    return tail
```

```python
def render(snek):
    s = 20
    h = int(s/2)
    board = np.zeros([s,s], dtype=str)
    for i,e in enumerate(snek[1:-1]):
        board[e[0]+h, e[1]+h] = f"{i}"
    board[snek[-1][0]+h, snek[-1][1]+h] = 'T'
    board[snek[0][0]+h, snek[0][1]+h] = 'H'
    print(np.rot90(board))
    #print(board)
```

```python
head = [0, 0]
tail = [0, 0]
tail_history = []
for direction, distance in moves:
    for _ in range(distance):
        move(head, direction)
        tail_move(head, tail)
        #render(head, tail)
        tail_history.append((*tail,))
len(set(tail_history))
```
